# Task 1 : Using Cross validation to improve the model

Divide the dataset into x_data and y_data parameters. Here y_data is the "Price" attribute, and x_data has all other attributes in the data set.


````python
y_data = df['Price']
x_data = df.drop('Price',axis=1)
````

Split the data set into training and testing subests such that you reserve 10% of the data set for testing purposes.

**test_size**: between 10-30% is common
* For small datasets: (e.g., less than a few thousand samples), a larger test set (20-30%) is preferred to ensure a sufficient sample size for evaluation.

* For larger datasets: (tens of thousands of samples), a smaller test size (10-20%) can be sufficient since the training set will still be large enough to build a robust model.

* High Variance Data: If the data has high variance or noise, you might prefer a larger test set (e.g., 25-30%) to get a better estimation of model performance.

* Low Variance Data: For more stable data, a smaller test set (10-20%) might suffice.


````python
# fixing random_state to a fixed quantity helps maintain uniformity between multiple
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10, random_state=1)
print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
````

Create a single variable linear regression model using "CPU_frequency" parameter.

Print the R^2 value of this model for the training and testing subsets.

Run a 4-fold cross validation on the model and print the mean value of R^2 score along with its standard deviation.


**1. Choosing cv Based on Dataset Size:**
* For small datasets (<100 samples), a higher value of cv is generally preferred. A common choice is Leave-One-Out Cross-Validation (LOOCV), where cv is equal to the number of samples. However, this can be computationally expensive.

* For Moderate-Sized datasets (=<10k samples), a 5-fold or 10-fold cross-validation (cv=5 or cv=10) is commonly used. This strikes a good balance between bias and variance in the model evaluation.

* For large datasets (>10k samples), even 3-fold or 5-fold cross-validation (cv=3 or cv=5) can be sufficient, as the large size already offers a robust estimate of model performance.

**2. Variance and Stability:**
* High Variance in Data: If your data has high variability or noise, using a higher cv (e.g., 10-fold) can give a more stable estimate of the model's performance.

* Low Variance in Data: For more homogeneous data, a lower cv (e.g., 3-fold or 5-fold) might be adequate and less computationally intensive.

**3. Typical Recommendations:**
* A standard starting point for most cases is 5-fold (cv=5) or 10-fold (cv=10) cross-validation. These values provide a good balance between computational cost and providing a robust estimate of the model's performance.

* When in Doubt: Use 10-fold cross-validation (cv=10), as it is a widely accepted standard for a variety of datasets and often yields reliable results.

````python
# Results from cross validation
Rcv = cross_val_score(lre, x_data[['CPU_frequency']], y_data, cv=4)
print(f"The mean R^2 of the folds is {Rcv.mean():.3f}, and the standard deviation is {Rcv.std():.3f}")
````

# Task 2: Overfitting

Split the data set into training and testing components again, this time reserving 50% of the data set for testing.


````python
# fixing random_state to a fixed quantity helps maintain uniformity between multiple
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.50, random_state=0)
print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
````

To identify the point of overfitting the model on the parameter "CPU_frequency", you'll need to create polynomial features using the single attribute.
You need to evaluate the R^2 scores of the model created using different degrees of polynomial features, ranging from 1 to 5. Save this set of values of R^2 score as a list.


````python
lre = LinearRegression()

Rsqu_test = []

order = [1, 2, 3, 4, 5]

for n in order:
    pr = PolynomialFeatures(degree=n)
    x_train_pr = pr.fit_transform(x_train[['CPU_frequency']])
    x_test_pr = pr.fit_transform(x_test[['CPU_frequency']])
    lre.fit(x_train_pr, y_train)
    Rsqu_test.append(lre.score(x_test_pr, y_test))
````

Plot the values of R^2 scores against the order. Note the point where the score drops.


````python
plt.plot(order, Rsqu_test)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
````

# Task 3 : Ridge Regression

Now consider that you have multiple features, i.e. 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU' and 'Category'. Create a polynomial feature model that uses all these parameters with degree=2. Also create the training and testing attribute sets.


````python
pr = PolynomialFeatures(degree=2)

Z = ['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']

x_train_pr = pr.fit_transform(x_train[Z])
x_test_pr = pr.fit_transform(x_test[Z])
````

<details><summary>Click here for the solution</summary>

```python

x_test_pr=pr.fit_transform(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
```

</details>


Create a Ridge Regression model and evaluate it using values of the hyperparameter alpha ranging from 0.001 to 1 with increments of 0.001.
Create a list of all Ridge Regression R^2 scores for training and testing data.


````python
# tqdm is a progress bar for loops, useful for long-running loops and visualize their progress.

Rsqu_test = []
Rsqu_train = []
Alpha = np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha)
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})
    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
````

Plot the R^2 values for training and testing sets with respect to the value of alpha


````python
width = 8
height = 6
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.ylim(0,1)
plt.legend()
````
